# CESD Results Analysis

Analyze POPE, CHAIR, and ablation results.

```python
import json
from pathlib import Path
import pandas as pd
```

```python
results_dir = Path("../results")
pope_files = list(results_dir.glob("pope*.json"))
chair_files = list(results_dir.glob("chair*.json"))
ablation_file = results_dir / "ablation" / "ablation_summary.json"
```

```python
# Load POPE results
pope_results = []
for f in pope_files:
    with open(f) as fp:
        d = json.load(fp)
    pope_results.append({"file": f.name, **d})
pd.DataFrame(pope_results)
```

```python
# Load ablation results
if ablation_file.exists():
    with open(ablation_file) as f:
        ablation = json.load(f)
    pd.DataFrame(ablation).T
```
